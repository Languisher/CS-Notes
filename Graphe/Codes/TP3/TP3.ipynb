### Exercice n°1

Créer une fonction `est_voisin(G,x,y)` qui prend entrée deux sommets et teste si ils sont voisins.

On peut changer les labels des sommets en utilisant un dictionnaire :

```python
def est_voisin(G, x, y):
    if G.adj[x]:
        for possible_adj in list(G.adj[x]):
            if y == possible_adj:
                return True 
        return False 
    return False

est_voisin(G, 0, 1)
```

### Exercice n°3

Soit $G$ un graphe connexe. Soit $x$ un sommet fixé. Soit $y$ un sommet. On note $d(y)$ la distance du sommet $y$ au sommet $x$.

**1)** Montrer que G est biparti **si et seulement si** il n'existe pas deux sommets reliés avec la même distance.

**2)** En déduire une méthode pour tester si un graphe est biparti. Le programmer.

### Correction exercice n°3

**1)** $\Rightarrow$. On raisonne par l'absurde.Supposons qu'il existe deux sommets reliés ayant la même distance, on les notera $y$ et $z$. Alors si ils sont reliés, ils ne sont pas dans la même partition. Supposons que $x$ est dans la même partition que $y$. Alors il faut un nombre paire d'arête pour aller de $x$ à $y$. Un nombre impair d'arêtes pour aller de $x$ à $z$. Ce qui est absurde.

$\Leftarrow$ Supposons qu'il n'existe pas deux sommets reliés ayant la même distance. On note :
$$A=\{y\in G,\quad d(y) pair\},\qquad B=\{z\in G,\quad d(z) impair\}.$$

Prenons deux éléments de $A$. Si il existe un chemin reliant deux point de $A$, alors il devrait y avoir une distance impaire. Il n'y a pas de chemin qui relie les éléments de $A$. Idem pour $B$.  Ainsi si deux sommets sont reliés, l'un est dans $A$, l'autre est dans $B$.

**2)** Pour tester si un graphe est biparti :
1. On choisit un sommet $x$.
1. On déterminer la distance de chacun des sommets à $x$.
1. On teste si deux sommets avec même distance sont reliés.
